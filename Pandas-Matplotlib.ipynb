```python
import pandas as pd
df=pd.read_csv('Ages.csv')
df.head()
```

```python
from matplotlib import pyplot as plt
plt.plot(df['Year'],df['Median Age of Males'],label='Males',color='seagreen',linewidth=2,linestyle='-',marker='s')
plt.plot(df['Year'],df['Median Age of Females'],label='Females',color='violet',linewidth=2,linestyle=':',marker='d')
plt.xlabel("Year",fontsize=15)
plt.ylabel("Age(Yrs)",fontsize=15)
plt.text(1950,20.5,"lowest age")
plt.title("Median Age of Males and Females",fontsize=18)
plt.style.use('seaborn')
plt.legend()
plt.show()
```
