```python
####### removing punctuation from address
```

```python
df=pd.read_csv('gradedata.csv')
df.head()
```

```python
import string
exclude=set(string.punctuation)
def remove_punctuation(x):
    try:
        x=''.join(ch for ch in x if ch not in exclude)
    except:
        pass
    return x
```

```python
df.address=df.address.apply(remove_punctuation)
```

```python
df
```

```python
def rem(x):
    print(x)
    x=''.join(ch for ch in x if ch not in exclude)
    return x
```

```python
df=pd.read_csv('gradedata.csv')
df.head()
```

```python
df_bins=df[['fname','lname','gender','age','hours','grade']]
```

```python
df_bins.head()
```

```python
bins=[0,60,70,80,90,100]
group_names=['F','D','C','B','A']
```

```python
df_bins['letter_grades']=pd.cut(df_bins['grade'],bins,labels=group_names)
```

```python
df_bins.head()
```

```python
df_bins.groupby('letter_grades')['hours'].mean()
```

```python
# count the number of students according to their grades
```
