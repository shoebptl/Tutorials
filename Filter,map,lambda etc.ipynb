# Case-2: Reduce and lambda

```python
nums=[10,20,30,40,50] 
# Add all the elements of the above list by using reduce function
```

```python
from functools import reduce
sum_list=reduce(lambda x,y:x+y,nums)
print(sum_list)
```

```python
# Find maximum
max_list=reduce(lambda x,y:x if x>y else y,nums)
print(max_list)
```

```python
# Find minimum
min_list=reduce(lambda x,y:x if x<y else y,nums)
print(min_list)
```

```python
users=pd.Series(['John A Smith','Jennefer M Brown','David D Moore'])
users
```

```python
def initials_1(name):
    initials=''
    name_list=name.split()
    for name in name_list:
        initials=initials+name[0]
    return initials
```

```python
initials_1('John A Smith')
```

```python
users.apply(initials_1)
```

```python
def initials_2(name):
    return ''.join([x[0] for x in name.split()])
```

```python
users.map(initials_2)
```

```python
users.map(initials_1)
```

```python
def grade(per):
    if per>=90:
        return 'A'
    elif per>=80 and per<90:
        return 'B'
    elif per>=70 and per<80:
        return 'C'
    else:
        return 'Pass'
```

```python
marks['Grade']=marks['Percentage'].apply(grade)
```
