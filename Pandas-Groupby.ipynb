```python
import pandas as np
fortune=np.read_csv('fortune1000.csv')
fortune.head()
```

```python
# Find 5 top sectors in terms of their Revenue
sector=fortune.groupby('Sector')
rev_sum=sector['Revenue'].sum()
rev_sum.sort_values(ascending=False).head(5)
```

```python
# Find 5 top sectors in terms of their Profits
profit_sum=sector['Profits'].sum()
profit_sum.sort_values(ascending=False).head(5)
```

```python
# Find 5 top sectors in terms of their Employees size
emp_size=sector['Employees'].sum()
emp_size.sort_values(ascending=False).head(5)
```

```python
# Find 5 top sectors in terms of the no of companies
sector_count=sector['Rank'].count()
sector_count.sort_values(ascending=False).head(5)
```

```python
for label,row in sector:
    
    if label=="Energy":
        print(len(row))
```
