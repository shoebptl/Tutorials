```python
def remove_tags(str1):
    result=re.compile(r'<.*?>')
    return result.sub('',str1)
```

```python
# checking the function with a html string
x="<a href=""http://twitter.com/download/iphone"" rel=""nofollow"">Twitter for iPhone</a>"
remove_tags(x)
```

```python
df_archive_clean['source']=df_archive_clean['source'].apply(lambda x : remove_tags(x))
```

## Tidying data

```python
# clubbing doggo,pupper,puppo and floofer in one column and delete the original 4 columns
```

```python
def select_dog_stage(row):
    if row['doggo']=='doggo':
        return 'doggo'
    elif row['floofer']=='floofer':
        return 'floofer'
    elif row['pupper']=='pupper':
        return 'pupper'
    elif row['puppo']=='puppo':
        return 'puppo'
    else:
        return 'None'
```

```python
df_archive_clean['dog_stage']=df_archive_clean.apply(lambda x: select_dog_stage(x),axis=1)
```
